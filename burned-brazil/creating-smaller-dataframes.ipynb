```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import matplotlib.dates as mdates
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
import time
```

```python
df2014 = pd.read_csv('../../dataset/data2014.csv', index_col = 'datahora', usecols = ['riscofogo', 'precipitacao', 'bioma', 'datahora', 'latitude', 'longitude'], parse_dates = True)
df2015 = pd.read_csv('../../dataset/data2015.csv', index_col = 'datahora', usecols = ['riscofogo', 'precipitacao', 'bioma', 'datahora', 'latitude', 'longitude'], parse_dates = True)
df2016 = pd.read_csv('../../dataset/data2016.csv', index_col = 'datahora', usecols = ['riscofogo', 'precipitacao', 'bioma', 'datahora', 'latitude', 'longitude'], parse_dates = True)
df2017 = pd.read_csv('../../dataset/data2017.csv', index_col = 'datahora', usecols = ['riscofogo', 'precipitacao', 'bioma', 'datahora', 'latitude', 'longitude'], parse_dates = True)
df2018 = pd.read_csv('../../dataset/data2018.csv', index_col = 'datahora', usecols = ['riscofogo', 'precipitacao', 'bioma', 'datahora', 'latitude', 'longitude'], parse_dates = True)
df2019 = pd.read_csv('../../dataset/data2019.csv', index_col = 'datahora', usecols = ['riscofogo', 'precipitacao', 'bioma', 'datahora', 'latitude', 'longitude'], parse_dates = True)
df2020 = pd.read_csv('../../dataset/data2020.csv', index_col = 'datahora', usecols = ['riscofogo', 'precipitacao', 'bioma', 'datahora', 'latitude', 'longitude'], parse_dates = True)
```

```python
dataframes = [df2014, df2015, df2016, df2017, df2018, df2019, df2020]
biomes = ['Amazonia', 'Caatinga', 'Pantanal', 'Mata Atlantica', 'Pampa', 'Cerrado']
```

```python
for dataframe in dataframes:
    for biome in biomes:
        
        df_yb = dataframe[(dataframe.bioma == biome) & (dataframe.riscofogo == 1)].copy()
        df_yb = df_yb.resample('W')[['riscofogo', 'precipitacao']].sum()
        year = str(dataframe.index[0].year)
        df_yb.to_csv('cleaned-datasets/timeseries{}{}.csv'.format(year, biome))
        
```
